```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import sympy
from sympy import *
import numpy 
import control 
```

```python
# Step
N = 1000
t = np.linspace(0,10,N)
u = np.zeros(N)
u[floor(N/4):]=1

plt.figure()
plt.plot(t,u,linewidth =3)
#plt.title("Step Response")
#plt.grid()
plt.xlabel("time (s)")
plt.ylabel("x(t)")
plt.axis([0, 10, -0.1, 1.1])
plt.show()
```

```python
# Sine 1
N = 1000
t = np.linspace(0,10,N)
u = np.sin(2*np.pi*0.5*t)


plt.figure()
plt.plot(t,u,linewidth =3)
#plt.title("Step Response")
#plt.grid()
plt.xlabel("time (s)")
plt.ylabel("x(t)")
plt.axis([0, 10, -4, 4])
plt.show()
```

```python
# Sine 2
N = 1000
t = np.linspace(0,10,N)
u = 5*np.sin(2*np.pi*2*t)


plt.figure()
plt.plot(t,u,linewidth =2)
#plt.title("Step Response")
#plt.grid()
plt.xlabel("time (s)")
plt.ylabel("x(t)")
#plt.axis([0, 10, -0.1, 1.1])
plt.show()
```
