```python
%matplotlib inline
%pylab inline
pylab.rcParams['figure.figsize'] = (12, 6)

#import mpld3
#mpld3.enable_notebook()

#import matplotlib
import numpy as np
#import sympy
#from sympy import *
from control import matlab
import control
import scipy
import matplotlib.pyplot as plt

```

```python
sys = control.ss("1. -2; 3. -4", "5.; 7", "6. 8", "9.")
mag, phase, omega = control.bode(sys)
```

```python
H = control.matlab.tf(1,1)
k = 7;
G = control.matlab.tf([0, 0, k],[1, 8, 0])
TFa = control.matlab.feedback(G,H)
y,t = control.matlab.step(TFa)
plt.figure()
plt.plot(t,y)
```

```python
Y = []
T = []
K = [7,16,80,400]
t = np.linspace(0, 5, 100)
for k in K:
    H = control.matlab.tf(1,1)
    G = control.matlab.tf([0, 0, k],[1, 8, 0])
    TF = control.matlab.feedback(G,H)
    y,t0 = control.matlab.step(TF,t)
    Y.append(y)
    T.append(t)
plt.figure()
plt.hold(True)
st = ['b-', 'k-','g-','y-']
for i, k in enumerate(K):
    plt.plot(T[i],Y[i],st[i],label='K= %d'%(k))
plt.grid(True)
plt.xlabel('Time t[s]')
plt.ylabel('y(t)')
plt.legend()
```
