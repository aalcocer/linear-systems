## RLC low pass filter time response 

<img src="images/RLClowpass.png" width="600" height="600">

<img src="images/RLClowpass2.png" width="500" height="400">




```python
def pltRLC(vin,L,R,C):
    N = 500
    t = np.linspace(0,10,N)
    A = -vin
    B = -vin/(R*C)
    a = 1/(2*R*C)
    b = np.sqrt(1/(L*C)-1/(4*(R**2)*(C**2)))
    c = 1/(L*C)
    
    vout = vin + np.exp(-a*t)*(A*np.cos(b*t)+((B-A*a)/b)*np.sin(b*t))
    plt.figure(figsize=(15,5))
    plt.plot(t,vout,linewidth=3,label='v_out(t)')
    plt.plot(t,vin*np.ones(N),linewidth=3,linestyle='--',label='v_in(t)')
    plt.xlabel('Time [s]')
    plt.ylabel('v_out [V]')
    plt.legend()
    plt.grid()
```

```python
ipywidgets.interact(pltRLC, vin=(0,20,0.1), L=(1e-1,2,0.1),R=(1,10,0.1),C=(1e-2,1,0.1));
```
